```python
def get_num_mem(x):
    temp=x.split('\u2009')
    return float(temp[0])
def get_num(x):
    temp=x.split(' ')
    return float(temp[0])
```

```python
df['Memory'].unique()
```

```python
def func(x):
    if '\u2009' in x:
        return get_num_mem(x)
    elif ' ' in x:
        return get_num(x)
def get_core(x):
    possible_cores=['i9','i7','i5','i3','9','5','7','3']
    temp=x.split(' ')
    for core in possible_cores:
        if core in temp:
            return core
    return None

def get_gen(x):
    possible_gen=[str(i)+"th" for i in range(15) if i!=0]
    temp=x.split(' ')
    for gen in possible_gen:
        if gen in temp:
            temp2=gen.split('t')
            gen_num=float(temp2[0]) 
            return gen_num
    return None 

```

```python
#in prices some are numbers some are integers
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def get_price(x):
    if not is_float(x):
        temp=x.split(',')
        price=''
        for part in temp:
            price+=part
        return float(price)
    else :
        return float(x)    
```

```python
model=LinearRegression()
params=model.get_params()
print(params)
```

```python
algo = {
    'linear_regressor':{
        'model':LinearRegression(),
        'params':
        {
            'fit_intercept': [True,False],
            'positive':[True,False]

        }
    },
    'lasso':
    {
        'model':Lasso(),
        'params':{
            'alpha':[1,2],
            'selection':['random','cyclic']
        }
    },
    'decision_tree':{
        'model':DecisionTreeRegressor(),
        'params':{
            'criterion':['squared_error','friedman_mse'],
            'splitter':['best','random']
        }
    }
}
```

```python
algo.items()
```

```python
def compute_best_gridsearchcv(X,y):
    scores=[]
    cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
    for algo_name,config in algo.items():
        gs=GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False,error_score='raise')  
        gs.fit(X,y)
        scores.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
compute_best_gridsearchcv(X,y) 
```
